## CO2 Equivalency

We're looking for 400 ppm CO2 at atmospheric conditions.

Use 298 K, 1.01325 bar as atmospheric

```python
# Partial Presure of CO2 at atmospheric conditions

p_CO2 = (400. / 1.0e6) * p_ATM
print('Partial Pressure of CO2 in atmosphere:', p_CO2, 'bar')
```

What about isotherm measurements at different temperature?

A sample of atmospheric gas, with a different temperature (T_1) would have:

p = (T_1/T_ATM) * p_ATM

assuming ideal gas properties. Note that I've looked up CO2 bulk PVT at 298 and the density departure function is < 1 % for p < 2 bar. Ideal gas EOS should be quite adequate, especially since we're just trying to bracket a pressure to search for.

Continuing... the mole fraction of CO2 is unchanged, so the partial pressure we're looking for is:

p_CO2 = (T_1/T_ATM) * p_ATM * (4e-4) = 0.00040530000000000004 bar * (T_1 / 298 K)

```python
# Example: isotherm measured at 350 K:

p_CO2_diff = 350. * p_CO2 / 298.
print(p_CO2_diff)
```
